## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# ==============================
# Data Wrangling
# ==============================

# Creating copies of original datasets
reviews_df = reviews.copy()
restaurants_df = restaurants.copy()

# ------------------------------
# Handling Missing Values
# ------------------------------

# Removing rows with missing review text and ratings
reviews_df.dropna(subset=['Review', 'Rating'], inplace=True)

# Filling missing values in categorical columns with 'Unknown'
reviews_df['Reviewer'].fillna('Unknown', inplace=True)
reviews_df['Metadata'].fillna('Unknown', inplace=True)
reviews_df['Time'].fillna('Unknown', inplace=True)

restaurants_df['Collections'].fillna('Not Available', inplace=True)
restaurants_df['Timings'].fillna('Not Available', inplace=True)

# ------------------------------
# Removing Duplicate Records
# ------------------------------

reviews_df.drop_duplicates(inplace=True)
restaurants_df.drop_duplicates(inplace=True)

# ------------------------------
# Converting Data Types
# ------------------------------

# Convert Rating column to numeric if required
reviews_df['Rating'] = pd.to_numeric(reviews_df['Rating'], errors='coerce')

# Remove rows having invalid ratings after conversion
reviews_df.dropna(subset=['Rating'], inplace=True)

# ------------------------------
# Reset Index
# ------------------------------

reviews_df.reset_index(drop=True, inplace=True)
restaurants_df.reset_index(drop=True, inplace=True)

# ------------------------------
# Final Shape of Datasets
# ------------------------------

print("Final Shape of Reviews Dataset:", reviews_df.shape)
print("Final Shape of Restaurants Dataset:", restaurants_df.shape)
````

### What all manipulations have you done and insights you found?

Data Manipulations Performed:

Created copies of the original datasets to preserve the raw data.
Handled missing values by removing rows with missing review text and ratings, as these fields are essential for sentiment analysis.
Replaced missing values in categorical columns such as Reviewer, Metadata, Time, Collections, and Timings with appropriate placeholder values like 'Unknown' and 'Not Available'.
Removed duplicate records from both datasets to avoid biased analysis and improve data quality.
Converted the Rating column into a numeric data type for further analysis and model building.
Reset the index after cleaning to maintain a continuous sequence of rows.

Insights Found:

The Reviews dataset initially contained missing values in multiple columns, indicating incomplete customer information.
A total of 36 duplicate records were found in the Reviews dataset, while the Restaurant dataset contained no duplicate records.
After data cleaning, the Reviews dataset contained 9,954 rows and 7 columns, while the Restaurant dataset contained 105 rows and 6 columns.
The datasets contain both textual and numerical features, making them suitable for sentiment analysis and machine learning applications.
Data preprocessing significantly improved the quality and reliability of the dataset for further analysis and predictive modeling.

````python
from textblob import TextBlob

# Chart - 6: Sentiment Distribution

# Function to classify sentiment based on rating
def get_sentiment_from_rating(rating):
    if rating >= 4:
        return 'Positive'
    elif rating == 3:
        return 'Neutral'
    else:
        return 'Negative'

# Apply the function to create a 'Sentiment' column
reviews_df['Sentiment'] = reviews_df['Rating'].apply(get_sentiment_from_rating)

plt.figure(figsize=(8,6))

sentiment_counts = reviews_df['Sentiment'].value_counts()

plt.pie(sentiment_counts.values,
        labels=sentiment_counts.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Distribution of Customer Sentiments')

plt.show()
````

There is no significant relationship between restaurant cost and customer ratings.

Alternative Hypothesis (H₁):

There is a significant relationship between restaurant cost and customer ratings.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

# Pearson Correlation Test
corr, p_value = pearsonr(merged_df['Cost'], merged_df['Rating'])

print("Correlation Coefficient:", corr)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
    print("There is a significant relationship between Cost and Rating.")
else:
    print("Fail to Reject Null Hypothesis")
    print("There is no significant relationship between Cost and Rating.")
````

##### Which statistical test have you done to obtain P-Value?

Pearson Correlation Test was used to obtain the p-value.

##### Why did you choose the specific statistical test?

The Pearson Correlation Test was selected because both Cost and Rating are numerical variables. This test helps measure the strength and direction of the linear relationship between these two variables and determines whether the relationship is statistically significant. It is appropriate when analyzing the association between two continuous numerical features.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Research Hypothesis

Statement: Positive reviews are significantly higher than negative reviews.

Null Hypothesis (H₀):

Positive and negative reviews occur in equal proportions.

Alternative Hypothesis (H₁):

Positive reviews occur significantly more frequently than negative reviews.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chisquare

# Count positive and negative reviews
sentiment_counts = reviews_df['Sentiment'].value_counts()

observed = sentiment_counts.values

# Assuming equal distribution as expected frequency
expected = [sum(observed)/2, sum(observed)/2]

# Perform Chi-Square Test
chi_stat, p_value = chisquare(observed, expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
    print("Positive and Negative reviews are not equally distributed.")
else:
    print("Fail to Reject Null Hypothesis")
    print("Positive and Negative reviews are equally distributed.")
````

##### Which statistical test have you done to obtain P-Value?

The Chi-Square Goodness of Fit Test was used to obtain the p-value.

##### Why did you choose the specific statistical test?

The Chi-Square Goodness of Fit Test was chosen because the analysis involves comparing the observed frequencies of categorical variables (Positive and Negative reviews) with their expected frequencies. This test helps determine whether the observed distribution differs significantly from the expected distribution and is appropriate for categorical data analysis.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Research Hypothesis

Statement: Different cuisines receive significantly different customer ratings.

Null Hypothesis (H₀):

There is no significant difference in the average customer ratings among different cuisines.

Alternative Hypothesis (H₁):

There is a significant difference in the average customer ratings among different cuisines.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Select top 5 cuisines based on frequency
top_cuisines = merged_df['Cuisines'].value_counts().head(5).index

# Create groups for ANOVA test
groups = []

for cuisine in top_cuisines:
    ratings = merged_df[merged_df['Cuisines'] == cuisine]['Rating']
    groups.append(ratings)

# Perform One-Way ANOVA Test
f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

# Conclusion
if p_value < 0.05:
    print("Reject Null Hypothesis")
    print("There is a significant difference in ratings among cuisines.")
else:
    print("Fail to Reject Null Hypothesis")
    print("There is no significant difference in ratings among cuisines.")
````

##### Which statistical test have you done to obtain P-Value?

The One-Way ANOVA (Analysis of Variance) Test was used to obtain the p-value.

##### Why did you choose the specific statistical test?

The One-Way ANOVA test was chosen because it is used to compare the mean ratings of more than two groups simultaneously. In this analysis, customer ratings are compared across multiple cuisine categories. ANOVA helps determine whether the differences in average ratings among cuisines are statistically significant.

#### 1. Expand Contraction

````python
# Expand Contraction
contractions = {
    "can't": "cannot",
    "won't": "will not",
    "n't": " not",
    "'re": " are",
    "'s": " is",
    "'d": " would",
    "'ll": " will",
    "'t": " not",
    "'ve": " have",
    "'m": " am"
}

def expand_contractions(text):
    for key, value in contractions.items():
        text = text.replace(key, value)
    return text

reviews_df['Review'] = reviews_df['Review'].apply(expand_contractions)
````

#### 2. Lower Casing

````python
# Lower Casing

reviews_df['Review'] = reviews_df['Review'].str.lower()
````

````python
# Remove Punctuations
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

reviews_df['Review'] = reviews_df['Review'].apply(remove_punctuation)
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
def remove_urls_digits(text):

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Remove words containing digits
    text = re.sub(r'\w*\d\w*', '', text)

    return text

reviews_df['Review'] = reviews_df['Review'].apply(remove_urls_digits)
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)

reviews_df['Review'] = reviews_df['Review'].apply(remove_stopwords)
````

The evaluation metrics considered for positive business impact were Accuracy, Precision, Recall, and F1-Score.

Accuracy: Measures the overall correctness of the model. A high accuracy indicates that the model can correctly classify customer sentiments, helping businesses understand customer opinions effectively.
Precision: Measures how many reviews predicted as positive are actually positive. High precision reduces false positive predictions and ensures that business decisions are based on reliable customer feedback.
Recall: Measures how many actual positive reviews are correctly identified. High recall helps businesses capture maximum customer feedback and ensures that important negative reviews are not missed.
F1-Score: It provides a balance between precision and recall. A high F1-score indicates that the model performs well in identifying customer sentiments without favoring one class over another.

These metrics collectively help businesses monitor customer satisfaction, identify service issues, and improve decision-making.
